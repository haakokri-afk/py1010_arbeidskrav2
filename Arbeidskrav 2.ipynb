```python
# Oppgave 5

import numpy as np

def figur(a, b):
    areal = (np.pi*a**2)/2 + (a*b)/2    # areal halvsirkel pluss areal rektangel. Lar feilen med diameter stå.
    omkrets = (2*np.pi*a)/2 + 2*(a+b)-a  # omkrets halvsirkel pluss omkrets rektangel minus a
    return (areal, omkrets)
    

lengde = float(input("Hva er rektangelets lengde/halvsirkelens diameter?"))
bredde = float(input("Hva er rektangelets bredde?"))
(fig_ar, fig_om) = figur(lengde, bredde)
print(f"Figurens areal = {fig_ar:.3f} cm")  # satser på cm, men figuren kan jo være i km for alt jeg vet
print(f"Figurens omkrets = {fig_om:.3f} cm")



    
```

```python
# Oppgave 6
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return -x**2 - 5 # Må innrømme at jeg ikke forstår oppgaven og formelen fullt ut. Trengte litt hjelp.
    
x = np.linspace(-10, 100, 200)    # oppretter array med 200 verdier fra -10 til 100.
y = f(x)                          # kjører funksjonen med input fra x i formelen

plt.plot(x, y)                    # aner ikke hva resultatet i plottet egentlig viser. 
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")



```
